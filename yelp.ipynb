```python
# Split the column ID and Address from the big data set to form a table in sqlite database
id_address = {}
for ad in search_result['businesses']:
    id_address[str(ad['id'])] = str(ad['location']['address1'])

df_address = pd.DataFrame(id_address.items(),columns=['id','address'])
```

```python
category = {}
for cat in search_result['businesses']:
    category[str(cat['id'])] = str(cat['categories'][0]['title'])

df_categories = pd.DataFrame(category.items(),columns=['id','categories'])
```

```python
import sqlite3
from sqlite3 import Error

def create_connection(path):
    connection = None
    try:
        connection = sqlite3.connect(path)
        print("Connection to SQLite DB successful")
    except Error as e:
        print(f"The error '{e}' occurred")

    return connection

#Create database for yelp data
connection = create_connection("yelpData.sqlite")
cursor = connection.cursor()
```

```python
#import data from DataFrame into SQLite database
df = df.applymap(str)
df.to_sql(name='restaurant',con=connection,if_exists="append")
```

```python
#add an address table into the database with ID
df_address.to_sql(name='address',con=connection,if_exists="append")
```

```python
#add a categories table into database with ID
df_categories.to_sql(name='categories',con=connection,if_exists="append")
```
